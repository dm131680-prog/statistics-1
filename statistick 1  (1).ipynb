Question 1: What is the difference between descriptive statistics and inferential
statistics? Explain with examples.

Answer ; The difference between descriptive statistics and inferential statistics lies in what they are used for and how far their conclusions extend.

1. Descriptive Statistics

Descriptive statistics are used to summarize and describe the main features of a dataset. They do not go beyond the data you already have.

Key points:
Focuses only on organizing, summarizing, and presenting data
Uses measures like:
Mean (average)
Median
Mode
Standard deviation
Charts/graphs (bar charts, pie charts)
Example:

Suppose a teacher records marks of 30 students in a class.

She calculates the average score = 65
Finds the highest = 90, lowest = 40
Draws a graph of marks distribution

This is descriptive statistics because it only describes the performance of those 30 students.

2. Inferential Statistics

Inferential statistics are used to make predictions or generalizations about a larger population based on a sample of data.

Key points:
Uses sample data to draw conclusions about a population
Involves probability
Includes:
Hypothesis testing
Confidence intervals
Regression analysis
Example:

A researcher studies 100 voters in a city to predict election results.

Finds that 60% support a candidate
Concludes that the candidate is likely to win in the entire city

This is inferential statistics because it uses a sample to make a broader prediction.

Question 2: What is sampling in statistics? Explain the differences between random
and stratified sampling.

Answer ; Sampling in Statistics

Sampling is the process of selecting a subset (sample) from a larger group (population) to study and draw conclusions about the whole population.

Example:

If a researcher wants to study the height of all students in a school (500 students), they may select 50 students and measure their heights.
This selected group is called a sample, and the process is sampling.

1. Random Sampling

Random sampling is a method where every member of the population has an equal chance of being selected.

Key features:
Selection is completely by chance
No bias in choosing participants
Simple and widely used
Example:
Writing all students’ names on slips and picking 50 randomly
Using a computer-generated random list
2. Stratified Sampling

Stratified sampling divides the population into subgroups (called strata) based on certain characteristics, and then samples are taken from each group.

Key features:
Population is divided into homogeneous groups
Samples are taken from each subgroup
Ensures representation of all categories
Example:

In a school:

Divide students into groups like boys and girls, or different classes
Then select students from each group proportionally


Differences Between Random and Stratified Sampling
Feature	Random Sampling	Stratified Sampling
Method	Selection by chance	Selection from subgroups
Population division	Not divided	Divided into strata
Representation	May miss some groups	Ensures all groups are included
Complexity	Simple	More complex
Accuracy	Less accurate if population is diverse	More accurate for diverse populations
In Simple Terms
Random Sampling = “Pick anyone randomly”
Stratified Sampling = “Divide into groups, then pick from each group”


Question 3: Define mean, median, and mode. Explain why these measures of central
tendency are important.

Answe ; ## **Measures of Central Tendency**

Measures of central tendency are statistical tools used to **identify the central or typical value** in a dataset. The three main measures are **mean, median, and mode**.

---

## **1. Mean (Average)**

The **mean** is the sum of all values divided by the total number of values.

### Formula:

[
\text{Mean} = \frac{\text{Sum of all observations}}{\text{Number of observations}}
]

### Example:

Data: 2, 4, 6, 8
Mean = (2 + 4 + 6 + 8) ÷ 4 = **5**

---

## **2. Median**

The **median** is the **middle value** when data is arranged in ascending or descending order.

### Steps:

* Arrange data in order
* If odd number of values → middle value
* If even number → average of two middle values

### Example:

* Data: 3, 5, 7 → Median = **5**
* Data: 2, 4, 6, 8 → Median = (4 + 6) ÷ 2 = **5**

---

## **3. Mode**

The **mode** is the value that appears **most frequently** in a dataset.

### Example:

Data: 2, 3, 3, 5, 7
Mode = **3**

---

## **Why These Measures Are Important**

### 1. Simplify Large Data

They reduce a large dataset into a **single representative value**, making it easier to understand.

### 2. Help in Comparison

Used to **compare different groups** (e.g., average marks of two classes).

### 3. Aid Decision-Making

Useful in business, economics, and research for making informed decisions.

### 4. Show Different Aspects of Data

* **Mean** → overall average
* **Median** → central position (useful when extreme values exist)
* **Mode** → most common value

---

## **In Simple Terms**

* **Mean** = Average
* **Median** = Middle value
* **Mode** = Most frequent value

---

Question 4: Explain skewness and kurtosis. What does a positive skew imply about the
data?
Answer ; ## **Skewness and Kurtosis**

### **1. Skewness**

Skewness refers to the **asymmetry of a distribution**—that is, whether the data is spread more to one side of the mean than the other.

### Types of Skewness:

* **Positive Skew (Right-skewed)**
* **Negative Skew (Left-skewed)**
* **Zero Skew (Symmetrical distribution)**

### Explanation:

* In a **symmetrical distribution**, the left and right sides are equal.
* In a **skewed distribution**, one tail is longer than the other.

---

### **2. Kurtosis**

Kurtosis measures the **peakedness or flatness of a distribution** compared to a normal distribution.

### Types of Kurtosis:

* **Leptokurtic** → Highly peaked (more extreme values)
* **Mesokurtic** → Normal peak (like a normal distribution)
* **Platykurtic** → Flat distribution (fewer extreme values)

---

## **What Does Positive Skew Imply?**

A **positive skew** (right-skewed distribution) means:

* The **tail on the right side is longer**
* Most of the data values are **concentrated on the left**
* A few **large values (outliers)** pull the mean to the right

### Relationship:

[
\text{Mean} > \text{Median} > \text{Mode}
]

### Example:

Income distribution in a society:

* Most people earn moderate incomes
* A few people earn extremely high incomes
  → This creates a **positive skew**

---

## **Why Skewness and Kurtosis Are Important**

* Help understand the **shape of data distribution**
* Identify **outliers and extreme values**
* Useful in **data analysis, finance, and research**
* Help choose appropriate statistical methods

---

## **In Simple Terms**

* **Skewness** = Direction of data spread (left or right)
* **Kurtosis** = Shape of the peak (sharp or flat)
* **Positive skew** = Data piled on the left, long tail on the right

Question 5: Implement a Python program to compute the mean, median, and mode of
a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
(Include your Python code and output in the code box below.)

Answer ;




```python
# Program to calculate Mean, Median, and Mode

# Given list

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# -------- Mean --------

mean = sum(numbers) / len(numbers)

# -------- Median --------

sorted_numbers = sorted(numbers)
n = len(sorted_numbers)

if n % 2 == 0:
    median = (sorted_numbers[n//2 - 1] + sorted_numbers[n//2]) / 2
else:
    median = sorted_numbers[n//2]

# -------- Mode --------

frequency = {}

for num in numbers:
    if num in frequency:
        frequency[num] += 1
    else:
        frequency[num] = 1

max_freq = max(frequency.values())

mode = []
for key in frequency:
    if frequency[key] == max_freq:
        mode.append(key)

# -------- Output --------

print("Mean =", mean)
print("Median =", median)
print("Mode =", mode)
```

Question 6: Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]
(Include your Python code and output in the code box below.)

Answer ;

```python
# Program to calculate Covariance and Correlation Coefficient

# Given data

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

n = len(list_x)

# Mean of X and Y

mean_x = sum(list_x) / n
mean_y = sum(list_y) / n

# Covariance

cov = 0
for i in range(n):
    cov += (list_x[i] - mean_x) * (list_y[i] - mean_y)

covariance = cov / n

# Standard Deviations

sum_x = 0
sum_y = 0

for i in range(n):
    sum_x += (list_x[i] - mean_x) ** 2
    sum_y += (list_y[i] - mean_y) ** 2

std_x = (sum_x / n) ** 0.5
std_y = (sum_y / n) ** 0.5

# Correlation Coefficient

correlation = covariance / (std_x * std_y)

# Output

print("Covariance =", covariance)
print("Correlation Coefficient =", correlation)
```

Question 7: Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
(Include your Python code and output in the code box below.)

Answer ;

```python
# Program to draw boxplot and find outliers

import matplotlib.pyplot as plt

# Given data

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# -------- Boxplot --------

plt.boxplot(data)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.show()

# -------- Finding Outliers --------

# Step 1: Sort data

data_sorted = sorted(data)
n = len(data_sorted)

# Step 2: Find Q1 and Q3

Q1 = data_sorted[n//4]
Q3 = data_sorted[(3*n)//4]

# Step 3: IQR

IQR = Q3 - Q1

# Step 4: Bounds

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 5: Outliers

outliers = []
for x in data_sorted:
    if x < lower_bound or x > upper_bound:
        outliers.append(x)

# Output

print("Q1 =", Q1)
print("Q3 =", Q3)
print("IQR =", IQR)
print("Lower Bound =", lower_bound)
print("Upper Bound =", upper_bound)
print("Outliers =", outliers)
```

Explanation
IQR (Interquartile Range) = 8
Acceptable range = 3 to 35
All values lie within this range
👉 So, no outliers are present

Question 8: You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
● Explain how you would use covariance and correlation to explore this
relationship.
● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]   
daily_sales = [2200, 2450, 2750, 3200, 4000]
(Include your Python code and output in the code box below.)

answer ; Using Covariance and Correlation (Concept Explanation)

As a data analyst, you can explore the relationship between advertising spend and daily sales using:

1. Covariance
Measures the direction of relationship
If covariance is:
Positive → as advertising increases, sales also increase
Negative → as advertising increases, sales decrease

👉 It tells direction, but not strength clearly.

2. Correlation
Measures both strength and direction of relationship
Value ranges from -1 to +1:
+1 → perfect positive relationship
0 → no relationship
-1 → perfect negative relationship

👉 It tells how strongly advertising affects sales

```python
# Program to calculate correlation between advertising spend and daily sales

# Given data

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

n = len(advertising_spend)

# Mean

mean_x = sum(advertising_spend) / n
mean_y = sum(daily_sales) / n

# Covariance

cov = 0
for i in range(n):
    cov += (advertising_spend[i] - mean_x) * (daily_sales[i] - mean_y)

covariance = cov / n

# Standard deviations

sum_x = 0
sum_y = 0

for i in range(n):
    sum_x += (advertising_spend[i] - mean_x) ** 2
    sum_y += (daily_sales[i] - mean_y) ** 2

std_x = (sum_x / n) ** 0.5
std_y = (sum_y / n) ** 0.5

# Correlation

correlation = covariance / (std_x * std_y)

# Output

print("Covariance =", covariance)
print("Correlation =", correlation)
```

Interpretation
Correlation ≈ 0.99 → Very strong positive relationship
This means:
👉 As advertising spend increases, daily sales also increase significantly

Question 9: Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.
● Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.
● Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]
(Include your Python code and output in the code box below.)

answer ; Understanding Customer Satisfaction Data
1. Summary Statistics to Use

To understand the distribution of customer satisfaction scores (1–10 scale), we use:

(a) Mean
Gives the average satisfaction level
Helps understand overall customer sentiment
(b) Median
Gives the middle value
Useful when data has extreme values
(c) Standard Deviation
Measures how spread out the responses are
Low value → customers mostly agree
High value → mixed opinions
(d) Mode
Shows the most common score
Helps identify most frequent customer experience
2. Visualizations to Use
(a) Histogram
Shows distribution of scores
Helps see whether ratings are skewed or balanced
(b) Boxplot (optional)
Shows median, quartiles, and outliers
Helps detect unusual responses

```python
# Program to draw histogram for survey data

import matplotlib.pyplot as plt

# Given data

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Histogram

plt.hist(survey_scores, bins=5, color='skyblue', edgecolor='black')

plt.title("Customer Satisfaction Distribution")
plt.xlabel("Survey Scores (1-10)")
plt.ylabel("Frequency")

plt.show()
```

Output (Graph Interpretation)
Most scores are between 7 and 9
Peak frequency around 7–8
Very few low ratings (4–5)
Overall satisfaction is moderately high
Conclusion
Customers are generally satisfied
Distribution is slightly concentrated toward higher scores
Before product launch, company can focus on improving low-score feedback areas
